```python
##A- TOMA Y VISUALIZACIÓN DE DATOS##

#1 Importar librerias
from google.colab import drive
import pandas as pd
from IPython.display import display

#2 Configurar Pandas para mejor visualización
pd.set_option('display.max_rows', None)  # Muestra todas las filas (puedes ajustar el valor)
pd.set_option('display.max_columns', None)  # Muestra todas las columnas
pd.set_option('display.width', 1000)  # Ajusta el ancho de la pantalla

#3 Montar Google Drive
drive.mount('/content/drive')

#4 Definir ruta del archivo Excel
ruta_del_archivo = "/content/drive/MyDrive/01-Nube-larambla.master/Fuentes Informes/Ventas por producto 2022-2024.xlsx"

#5 cargar/leer el archivo
datos_ventas = pd.read_excel(ruta_del_archivo)

#6 Crear una copia del DataFrame original
datos_ventas_mod = datos_ventas.copy()

#7 Visualizar/mostrar las primeras filas de manera más ordenada
display(datos_ventas_mod.head(20))  # Muestra las primeras 20 filas

#8 Eliminar columnas que no necesitas del Data Frame (corregir errores)
datos_ventas = datos_ventas_mod.drop(columns=['CodigoBarra', 'Rubro','Marca'],inplace=True)

# Meter el parametro ",inplace=True" actualizaria el DF, pero es poco seguro#
#Reasignar y modificar DF son dos cosas distintas#

##B- DEPURAR DATAFRAME "DATOS VENTAS MOD"
#Eliminar con "dropna" todos los vacios

datos_ventas_mod = datos_ventas_mod.dropna(how='all')
datos_ventas_mod = datos_ventas_mod[~datos_ventas_mod['Total Vta.'].str.contains('Página', na=False)]


# Definir los valores que corresponden al encabezado (los que quieres excluir)
encabezado = ['Código', 'Artículo', 'Total Unid.', 'Total Vta.', '2024']

# Eliminar las filas que tengan exactamente esos valores
datos_ventas_mod = datos_ventas_mod[datos_ventas_mod['Código'] != 'Código']


# Verificar los cambios
display(datos_ventas_mod.head(20))

datos_ventas_mod.to_excel("/content/drive/MyDrive/01-Nube-larambla.master/Fuentes Informes/Ventas por producto 2022-2024_mod.xlsx", index=False)

##EDA (Exploratory Data Analysis)

```
